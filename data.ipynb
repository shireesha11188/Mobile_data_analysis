```python

import pandas as pd
from sqlalchemy import create_engine

server="LAPTOP-91GGJ4MU"
database="MOBILE"
driver= "ODBC Driver 17 for SQL Server"

engine = create_engine(f"mssql+pyodbc://@{server}/{database}?trusted_connection=yes&driver={driver}")
df=pd.read_csv(r"C:\Users\HP\Desktop\excel files\mobile_usage.csv")
```

```python
column_mapping={
    'Phone Brand':'Brand',
    'Number of Apps Installed': 'App Count',
    'Social Media Time (hrs/day)': 'Social Time',
    'Monthly Recharge Cost (INR)': 'Monthly Recharge',
    
}
df.rename(columns=column_mapping,inplace=True)
```

```python
df.columns=df.columns.str.lower().str.replace(r"\s*\(.*?\)", "", regex=True)
df.columns = df.columns.str.replace(" ", "_")
```

```python
df.index=df.index+1
```

```python
df.to_sql("mobile_usage", engine, if_exists="replace", index=False)
```
